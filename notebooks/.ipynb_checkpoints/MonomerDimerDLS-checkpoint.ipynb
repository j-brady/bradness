## Quantifying dodecamer concentrations of PAN by DLS

In DLS the intensity of light scattering is proportional to the molecular weight squared multiplied by the concentration of the scatterer. The diffusion constant measured is therefore a weighted average.

$D_z = \frac{\sum_{i} M^2_i c_i D_{T,i}}{\sum_{i} M^2_i c_i}$

$R_h$ is calculated according to stokes einstein.

$R_h = \frac{k_B T}{6 \pi \eta D_T}$

The average $R_h$ is therefore

$R_{avg} = \frac{\sum_{i} M^2_i c_i}{\sum_{i} M^2_i c_i /R_i}$

In the case of a hexamer to dodecamer

$H + H <--> D$

And $ K_d = \frac{[H]^2}{[D]} $

Assuming two states (hexamer, dodecamer) then the concentration of dodecamers can be calculated 

$[dodecamer] = \frac{1-\chi_H}{\phi_D(\chi_D-1)+2\phi_H(1-\chi_H)} $

Where,

$\chi_H = \frac{R_{avg}}{R_{H,hexamer}}$,
$\chi_D = \frac{R_{avg}}{R_{H,dodecamer}}$,
$\phi_H = \frac{M^2_{hexamer}}{M^2_{hexamer} c_{total}}$,
$\phi_D = \frac{M^2_{dodecamer}}{M^2_{hexamer} c_{total}}$

With $M$ being molar mass of the molecule and $c_{total}$ being total protein concentration ($c_{total} = [H] + 2[D]$)

### Note

1. The $R_{avg}$ equation was rearranged to

$[dodecamer] = \frac{M_{hexamer}^2 c_{total} - \frac{R_{avg}  M_{hexamer}^2 c_{total} }{ R_{hexamer}} }{\frac{R_{avg}M_{dodecamer}^2 }{ R_{dodecamer}} - \frac{2 R_{avg}M_{hexamer}^2}{R_{hexamer}} - M_{dodecamer}^2 + 2 M_{hexamer}^2}$

2. We are assuming that $R_h$s of ~7.3 and ~13 nm correspond to 100% monomer and 100% dimer, respectively


```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from lmfit.models import LinearModel
%matplotlib inline
from dls.dls import stokes_einstein, viscosity
```

```python
data = pd.read_csv("190316_pan_pH9_conc_titration_melt_2_fits.csv")
data.head()
```

```python
data = data.drop(610)

total = 70.
vols = [70.,50.,30.,20.,10.,5.,2.,1.]
conc = 100.
concs = {}
num = 1

for v in vols:
    for i in range(3):
        concs[f"C{num}"] = (v / total) * conc
        num += 1
        
data["well"] = data.col.apply(lambda x: x.split("_")[0])
data["temperature"] = data.col.apply(lambda x: float(x.split("_")[1][:-3]))
data["conc"] = data.apply(lambda x: concs[x.well] * 1e-6,axis=1) # uM
data["Rh"] = data.apply(lambda x: stokes_einstein(x.D,viscosity(x.temperature+273.15),x.temperature+273.15),axis=1)
```

```python
class MonomerDimer:
    
    def __init__(self,M_h=1.,M_d=2.,C_tot=100.e-6,R_h=7.3e-9,R_d=13.e-9):
        """ class for extracting monomer-dimer populations from DLS data 
        
            Arguments:
                M_h   -- mass of monomer
                M_d   -- mass of dimer
                C_tot -- total concentration of protein
                R_h   -- expected R_h of monomer
                R_d   -- expected R_h of dimer
        
        """
        self._M_h = M_h
        self._M_d = M_d
        self._C_tot = C_tot
        self._R_h = R_h
        self._R_d = R_d
        
    @property
    def M_h(self):
        return self._M_h
    
    @property
    def M_d(self):
        return self._M_d
    
    @property
    def C_tot(self):
        return self._C_tot

    @property
    def R_h(self):
        return self._R_h
    
    @property
    def R_d(self):
        return self._R_d
    
    def dimer(self, r_avg):
        """ Calculate dimer concentration 
            
            Arguments:
            r_avg -- observed r_h
            
            Returns:
            concentration of dimer
            
        """
        chi_h = r_avg / self.R_h
        chi_d = r_avg / self.R_d
        phi_h = self.M_h**2. / (self.M_h**2. * self.C_tot)
        phi_d = self.M_d**2. / (self.M_h**2. * self.C_tot)
        return (1. - chi_h) / (phi_d * (chi_d - 1.) + 2. * phi_h * (1. - chi_h) )
    
    def monomer(self, r_avg):
        """ Calculate monomer concentration 
        
            Arguments:
            r_avg -- observed r_h
            
            Returns:
            concentration of monomer
        """
        conc_dimer = self.dimer(r_avg)
        return self.C_tot - 2.*conc_dimer

# calculation for PAN at 60C 300 uM at pH 9.5
pan_300 = MonomerDimer(C_tot=300e-6)
pan_300.monomer(8.7e-9) / pan_300.C_tot
```

# PAN at pH 9.5 - 75 mM NaCl, 25 mM borate
For PAN at 300 $\mu$M (protomer) at pH 9.5 the expected fraction of hexamer in the sample would be ~78% at 60 degrees (see above).

```python
monomer_dimer = MonomerDimer(C_tot=100e-6) 
rhs = np.linspace(7.3e-9,13.e-9)
fraction = (monomer_dimer.monomer(rhs)) / monomer_dimer.C_tot
plt.plot(rhs*1e10,fraction,"-",label="monomer")
#plt.plot(monomer_dimer.monomer(rhs),rhs,"-",label="monomer")
plt.ylabel("Fraction of monomer")
plt.xlabel("$R_h [\AA]$")
plt.grid()
```

```python
monomer_dimer = MonomerDimer()
data["dodecamer"] = data.apply(lambda x: MonomerDimer(C_tot=x.conc).dimer(x.Rh), axis=1)
data["monomer"] = data.apply(lambda x: MonomerDimer(C_tot=x.conc).monomer(x.Rh), axis=1)
data["monomer_fraction"] = data.monomer / data.conc 
data["K"] = data.dodecamer.apply(lambda x: (100. - 2*x)**2/x)
#data.K.head()
```
