$D_z = \frac{\sum_{i} M^2_i c_i D_{T,i}}{\sum_{i} M^2_i c_i}$

$R_h$ is calculated according to stokes einstein.

$R_h = \frac{k_B T}{6 \pi \eta D_T}$

The average $R_h$ is therefore

$R_{avg} = \frac{\sum_{i} M^2_i c_i}{\sum_{i} M^2_i c_i /R_i}$

```python
def dimer(Kd,Ctot):
    return (Kd + 4*Ctot - np.sqrt(Kd**2. + 8.*Kd*Ctot)) / 8.

def Ravg(M,c,Rh):
    return np.sum(M**2.*c, axis=1)/np.sum((M**2.*c) / Rh, axis=1)
```

For three states,

$Ctot = M + 2D + 4T$

$Kd_1 = \frac{M^2}{D}$

$Kd_2 = \frac{D^2}{T}$

$0 = 4\frac{\left(\frac{M^2}{Kd_1}\right)^2}{Kd_2} + 2\frac{M^2}{Kd_1} + M - Ctot$

$0 = ax^4 + bx^3 + cx^2 + dx + e$

$a = 4\frac{Kd_2}{Kd_1^2}$,

$b = 0$,

$c = \frac{2}{Kd_1}$,

$d = 1$,

$e = -Ctot$

```python
M_m = 1.
M_d = 2.
M_t = 4.
R_m = 7.
R_d = 12.
R_t = 14.

Ctot = np.linspace(1e-6,200e-6)
Kd_1 = 50e-6
Kd_2 = 50e-6
```

```python
a = 4*(Kd_2/Kd_1**2.)
b = 0
c = 2./Kd_1
d = 1
e = -Ctot
M = np.real(np.array([np.roots([a,b,c,d,-e_i]) for e_i in Ctot])[:,3])
D = M**2. / Kd_1
T = D**2. / Kd_2
```

```python
plt.plot(Ctot, M, label="M")
plt.plot(Ctot, D, label="D")
plt.plot(Ctot, T, label="T")
plt.legend()
```

```python
c = np.vstack(zip(M,D,T))
M = np.array([M_m, M_d, M_t])
Rh = np.array([R_m, R_d, R_t])
R_avgs = Ravg(M,c,Rh)
plt.plot(Ctot,R_avgs)
```

```python
wells = [
    ("F",[13,14,15],2.71),
    ("F",[16,17,18],5.42),
    ("F",[19,20,21],8.13),
    ("F",[22,23,24],10.84),
    ("G",[1,2,3],16.26),
    ("G",[4,5,6],21.68),
    ("G",[7,8,9],27.1),
    ("G",[10,11,12],54.2),
    ("G",[13,14,15],135.5),
    ("G",[16,17,18],271.),
]
```

```python
data["well"] = data.col.apply(lambda x: x.split("_")[0])
data["temperature"] = data.col.apply(lambda x: float(x.split("_")[1][:-4]))
data["conc"] = 0.0

for v in wells:
    #print([nu for nu in v[0]])
    w = [f"{v[0]}{nu}" for nu in v[1]]
    #print(wells)
    conc = v[2]
    data.loc[data.well.isin(w),"conc"] = conc

data["Rh"] = data.apply(lambda x: stokes_einstein(x.D,viscosity(x.temperature+273.15),x.temperature+273.15),axis=1)

data = data[data.temperature<=48]
data.tail()
```

```python
Ts = []
concs = []
Rhs = []
Rh_errs = []
lens = []

for ind, row in data.groupby("temperature"):
    
    #plt.plot(row.conc,row.Rh,"o")
    means = row.groupby("conc").mean()
    stds = row.groupby("conc").std()
    #print(means.Rh,stds.Rh,means.temperature)
    length = len(means.index)
    lens.append(length)
    concs.extend(means.index*1e-6)
    Rhs.extend(means.Rh)
    Rh_errs.extend(stds.Rh)
    Ts.extend(means.temperature+273.15)
    
    plt.errorbar(means.index,means.Rh,yerr=stds.Rh,fmt="o")
    
print(np.array(concs).shape)
print(np.array(Rhs).shape)
print(np.array(Ts).shape)
```

```python
def Ravg(M,c,Rh):
    return np.sum(M**2.*c, axis=2)/np.sum((M**2.*c) / Rh, axis=2)

def mod(CT, R_1, R_3, R_6, R_12, dH1, dH2, dH3, dS1, dS2, dS3):
    
    R_1=3.e-9
    R_3=5.e-9 
    R_6=6.e-9 
    R_12=10.e-9
    
    R = 8.314
    
    C = CT[:,0]
    T = CT[:,1]

    M1  = 1.
    M3  = M1 * 3.
    M6  = M3 * 2.
    M12 = M6 * 2.
    

    #print(C.shape)
    #print([m for m in C])
    C_1 = []
    for m_i,T_i in CT:
        
        #print(m_i,T_i)
        
        Kd1 = np.exp(-dH1/(R*T_i) + dS1/R)
        print(dH1, dS1, Kd1)
        Kd2 = np.exp(-dH2/(R*T_i) + dS2/R)
        print(dH2, dS2, Kd2)
        Kd3 = np.exp(-dH3/(R*T_i) + dS3/R)
        print(dH3, dS3, Kd3)

        a = 12. / (Kd3 * Kd2**2. * Kd1**4.)
        b = 0
        c = 0
        d = 0
        e = 0
        f = 0
        g = 6. / (Kd2 * Kd1**2.)
        h = 0
        i = 0
        j = 3. / Kd1
        k = 0
        l = 1.
        print(m_i,a,g,j)
        roots = np.roots([a, b, c, d, e, f, g, h, i, j, k, l, -m_i])
        print("Roots",roots)
        roots = np.real(roots[np.where((np.imag(roots)==0)&(np.real(roots)>0))])
        print("Roots",roots)
#         print(roots)
        C_1.append(roots)
        
    C_1 = np.array(C_1)
    #print(C_1.shape)


        
#     C_1 = np.array(C_1)
#     C_1 = np.array([np.roots([a, b, c, d, e, f, g, h, i, j, k, l, -m]) for m in C])
#     print(C_1)
#     C_1 = np.real(C_1[np.where((np.imag(C_1)==0)&(np.real(C_1)>0))])
    
    C_3  = C_1**3. / Kd1
    C_6  = C_3**2  / Kd2
    C_12 = C_6**2  / Kd3

    Cs = np.array([C_1, C_3, C_6, C_12]).T
    #print(Cs.shape)
    #print(Cs)
    M  = np.array([M1, M3, M6, M12])
    #print(M.shape)
    Rh = np.array([R_1, R_3, R_6, R_12])
    #print(Rh.shape)
    R_avgs = Ravg(M,Cs,Rh)
    #print(R_avgs)
    return R_avgs
    
```

```python
CT = np.vstack(zip(concs,Ts))

Ravgs = np.array(Rhs)

model = Model(mod)
#params = model.make_params(R_1=3.e-9, R_3=5.e-9, R_6=6.e-9, R_12=10.e-9, dH1=7e4, dH2=-3e5, dH3=9e4, dS1=450., dS2=-850., dS3=400.)
params = model.make_params(R_1=3.e-9, R_3=5.e-9, R_6=6.e-9, R_12=10.e-9, dH1=10., dH2=10., dH3=1., dS1=1., dS2=1., dS3=1.)
out = model.fit(Ravgs,CT=CT, params=params)
print(out.fit_report())
```
